```python
import lib.graph
import lib.plot
```

```python
filename = '4932.protein.links.v11.5'
```

```python
graph = lib.graph.read_graph(filename)

thresholds = list(range(800, 1000, 50))
degree_sequences = [
    lib.graph.get_degree_sequence(
        lib.graph.remove_edges_below_threshold(graph, threshold) # 
    ) for threshold in thresholds
]
max_degree_sequence = max(degree_sequences[-1])

lib.plot.plot_histograms_n_by_2(
    xs=degree_sequences, 
    num_bin=100, 
    xscale='log', 
    yscale='log', 
    titles=thresholds, 
    xlims=[max_degree_sequence for i in thresholds] # Align to same xlim
)
```

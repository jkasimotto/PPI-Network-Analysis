```python
import itertools
import networkx as nx

def flatten(nested_lists):
    return itertools.chain.from_iterable(nested_lists)

# Load the network
florentine = nx.florentine_families_graph()

# Q1
nx.draw(florentine, with_labels=True)

# Q2
is_cc = nx.is_connected(florentine)
```

### Question 3
The Medici family is remarkable amongst the four cut-families because it has extreme centrality compared to the other three.
This is visible in the graph and is confirmed by networkx's betweenness centrality function. Betweeness centrality uses shortest paths which implies for one family to reach another family, they often had to pass through the Medici family.

```python
cut_vertices = list(flatten(family for family in nx.all_node_cuts(florentine)))

centralities = nx.betweenness_centrality(florentine)
centralities_of_cut_vertices = [centralities[vertex] for vertex in cut_vertices]

import matplotlib.pyplot as plt
plt.bar(cut_vertices, centralities_of_cut_vertices)
```

```python

import matplotlib.pyplot as plt
% matplotlib inline
font = {'family': 'DejaVu Sans',
        'weight': 'bold',
        'size': 32}
plt.rc('font', **font)
```

```python
import lib.files
```

```python
complexes = lib.files.read_yhtp2008_csv()
```

```python
import itertools

proteins = list(itertools.chain.from_iterable(complexes))
protein_complex_counts = []
for protein in proteins:
    protein_complex_counts.append(0)
    for complex in complexes:
        if protein in complex:
            protein_complex_counts[-1] += 1
```

```python
bins = list(range(1, max(protein_complex_counts)))
plt.hist(sorted(protein_complex_counts), density=True, bins=[1,2,3,4,5])
plt.title("Number of complexes per protein")
plt.xticks([1,2,3,4,5])
plt.xlabel("Number of complexes")
plt.ylabel("% of proteins")
```
